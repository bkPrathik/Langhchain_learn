```python
# pip install langchain-groq langchain-core

from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
import os
```

```python
# Initialize model
llm = ChatGroq(
    api_key=os.getenv("GROQ_API_KEY"),
    model="openai/gpt-oss-120b",
    temperature=0.7,
)

# Chat history (acts as memory)
history = [
    SystemMessage(content="You are a helpful assistant.")
]

def chat(user_input: str) -> str:
    history.append(HumanMessage(content=user_input))  # add user msg
    response = llm.invoke(history)                     # send full history
    history.append(AIMessage(content=response.content)) # save AI reply
    return response.content
```

```python
# Run chat loop
if __name__ == "__main__":
    print("Chat started! Type 'quit' to exit.\n")
    while True:
        user = input("You: ")
        if user.lower() == "quit":
            break
        reply = chat(user)
        print(f"AI: {reply}\n")
```
